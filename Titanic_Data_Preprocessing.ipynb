```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/softwareWCU/Data-Preprocessing-for-ML-using-Titanic-Dataset/main/titanic2.csv")

df.head()
```

```python
# Handling missing value

df.drop(columns=['Cabin'], inplace=True)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mode())
```

```python
df.isnull().sum()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mode())
```

```python
df.isnull().sum()
```

```python
df['Fare'].describe()
```

```python
df['Fare'].fillna(df['Fare'].median(), inplace=True)
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# List of numeric columns to check for outliers
num_cols = ['Age']

for col in num_cols:
    # Step 1: Calculate Q1, Q3, and IQR
    Q1 = df[col].quantile(0.25)        # 25th percentile
    Q3 = df[col].quantile(0.75)        # 75th percentile
    IQR = Q3 - Q1                       # Interquartile range

    # Step 2: Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Step 3: Filter out outliers
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
sns.boxplot(data=df['Age'])   # Correct column selection
plt.title('Box plot of Age')  # Add title
plt.show()                    # Display the plot
```

```python
# See all unique (different) values in the Pclass column
print(df['Pclass'].unique())
```

```python
df['Pclass'] = df['Pclass'].str.replace('st', '', regex=False)
df['Pclass'] = df['Pclass'].str.replace('nd', '', regex=False)
df['Pclass'] = df['Pclass'].str.replace('rd', '', regex=False)
df['Pclass'] = df['Pclass'].str.strip()  # remove extra spaces
```

```python
df['Pclass'] = df['Pclass'].astype(int)
```

```python
df['Sex'].unique()
```

```python
df['Sex'] = df['Sex'].str.lower()
```

```python
df['Sex'] = df['Sex'].str.strip()
```

```python
df['Sex'].unique()
```

```python
df['Embarked'].unique()
```

```python
df['Embarked'] = df['Embarked'].str.lower()
```

```python
df['Embarked'] = df['Embarked'].str.strip()
```

```python
df['Embarked'] = df['Embarked'].replace({
    'southampton': 's',
    'cherbourg': 'c',
})
```

```python
df['Embarked'].unique()
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df['Sex'].unique()
```

```python
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')
```

```python
df = pd.concat([df, embarked_dummies], axis=1)
```

```python
df.drop('Embarked', axis=1, inplace=True)
```
